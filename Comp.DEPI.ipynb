# Features Engineering 

```python
df["cholratio"] = df["HDL"] / df["LDL"]
#AVG OF HDL , LDL
```

```python
df.head()
```

```python
df["BMI"] = df["weight(kg)"] / ((df["height(cm)"] / 100) ** 2)
```

```python
df.head()
```

```python
df["average_BP"] = (df["systolic"] + df["relaxation"]) / 2
# AVG OF BP
```

```python
conditions = [
    df['cholratio'] > 0.7,  #Excellent
    (df['cholratio'] > 0.5) & (df['cholratio'] <= 0.7),  #Good
    (df['cholratio'] > 0.3) & (df['cholratio'] <= 0.5),  # Medium
    df['cholratio'] <= 0.3  #danger
]
labels=['Excellent','Good','Medium','Danger']
df['chol_status'] = np.select(conditions, labels)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

# الأعمدة اللي هنطبق عليها اللوج
cols = ['Gtp', 'ALT', 'hemoglobin', 'age', 'height(cm)', 'weight(kg)',
        'waist(cm)', 'triglyceride', 'HDL', 'serum creatinine']

# نسخ الأعمدة وتطبيق log1p
x_log1p_df = np.log1p(df[cols])
x = x_log1p_df.values
y = df['smoking']

# تقسيم البيانات
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0, test_size=0.35)

# Scaling
s = StandardScaler()
xtrs = s.fit_transform(xtrain)
xtes = s.transform(xtest)

# نموذج Logistic Regression
model = LogisticRegression(max_iter=1000)  # زودنا max_iter عشان يضمن converge
model.fit(xtrs, ytrain)

# التنبؤ والتقييم
y_pred = model.predict(xtes)
accuracy = accuracy_score(ytest, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# تقسيم البيانات
y = df['smoking']
x = df.iloc[:, :-1]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0, test_size=0.35)

# تطبيق StandardScaler
s = StandardScaler()
xtrs = s.fit_transform(xtrain)
xtes = s.transform(xtest)

# نموذج Random Forest مع Regularization
model_rf = RandomForestClassifier(
    n_estimators=300,       # عدد الأشجار
    random_state=0,         
    max_depth=10,           # الحد الأقصى لعمق الشجرة
    min_samples_split=5,    # الحد الأدنى للعينات لتقسيم العقدة
    min_samples_leaf=4,     # الحد الأدنى للعينات في كل ورقة
    max_features='sqrt'     # استخدام الجذر التربيعي لعدد الميزات
)

# تدريب النموذج
model_rf.fit(xtrs, ytrain)

# التنبؤ
y_pred_rf = model_rf.predict(xtes)

# تقييم النموذج
accuracy_rf = accuracy_score(ytest, y_pred_rf)
print("Random Forest Accuracy:", round(accuracy_rf, 4))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# تقسيم البيانات
y = df['smoking']
x = df.iloc[:, :-1]
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0, test_size=0.35)

# تطبيق StandardScaler
s = StandardScaler()
xtrs = s.fit_transform(xtrain)
xtes = s.transform(xtest)

# نموذج Logistic Regression مع Regularization
model_lr = LogisticRegression(
    random_state=0,         # ثابت لتكرار النتائج
   max_iter=100,          # عدد التكرارات لتدريب النموذج
    solver='lbfgs',         # طريقة حل المعادلات (يمكنك تغييرها حسب الحاجة)
    C=1.0                   # معامل التنظيم (لتقليل التعقيد)
)

# تدريب النموذج
model_lr.fit(xtrs, ytrain)

# التنبؤ
y_pred_lr = model_lr.predict(xtes)

accuracy_lr = accuracy_score(ytest, y_pred_lr)
print("Logistic Regression Accuracy:", accuracy_lr)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# x = df[['Gtp','ALT','hemoglobin','age','height(cm)','weight(kg)','waist(cm)','triglyceride','HDL','serum creatinine']].values
x = x_log1p_df.values
y = df['smoking']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0, test_size=0.35)

# تقييس البيانات
s = StandardScaler()
xtrs = s.fit_transform(xtrain)
xtes = s.transform(xtest)

# نموذج Random Forest
model = RandomForestClassifier()
model.fit(xtrs, ytrain)

# التنبؤ
y_pred = model.predict(xtes)

# حساب الدقة
accuracy = accuracy_score(ytest, y_pred)
print("Accuracy:", accuracy)

# حساب F1 Score
f1 = f1_score(ytest, y_pred)
print("F1 Score:", f1)
```
