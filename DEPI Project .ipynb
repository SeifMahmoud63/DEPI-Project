```python
# order_purchase_timestamp → وقت الشراء 
#######################################################################
# order_approved_at → وقت الموافقة على الطلب ✅
# 📌 بعد ما الزبون يعمل الطلب، الموقع لازم يوافق عليه (مثلاً يتأكد من الدفع).
# 🔹 مثال: الطلب اتوافق عليه يوم 1 يناير 2025 الساعة 10:05 صباحًا.
# 📌 القيمة في البيانات: 2025-01-01 10:05:00
#######################################################################
# 3️⃣ order_delivered_carrier_date → وقت تسليم الطلب لشركة الشحن 🚚
# 📌 ده الوقت اللي فيه الموقع سلَّم الطلب لشركة الشحن عشان تبدأ التوصيل.
# 🔹 مثال: الطلب طلع من المخزن يوم 2 يناير 2025 الساعة 3:00 مساءً.
# 📌 القيمة في البيانات: 2025-01-02 15:00:00
#######################################################################
#  order_delivered_customer_date → وقت وصول الطلب للزبون 🎁
# 📌 ده الوقت اللي فيه الزبون استلم الطرد بتاعه فعليًا.
# 🔹 مثال: أحمد استلم الموبايل يوم 5 يناير 2025 الساعة 2:00 ظهرًا.
# 📌 القيمة في البيانات: 2025-01-05 14:00:00
#######################################################################
# 5️⃣ order_estimated_delivery_date → التاريخ المتوقع للتوصيل 🗓️
# 📌 ده موعد التوصيل المتوقع اللي ظهر للزبون وقت ما كان بيطلب المنتج.
# 🔹 مثال: لما أحمد طلب الموبايل، الموقع قال له إن التوصيل هيكون يوم 6 يناير 2025.
# 📌 القيمة في البيانات: 2025-01-06 00:00:00
#######################################################################
#shipping_limit_date
# ده التاريخ المفروض البايع يديه لشركه الشحن للوصول اخرو يسلم امتي

# دول هيتفكلهم اليوم بس
```
